## 匯入套件

```python
import pandas as pd
import json
import codecs
from pandas import json_normalize
from ckiptagger import WS, POS, NER
import csv
from gensim import corpora
from gensim import models
from gensim.models import word2vec
import os 
```

## 資料夾內容，存成list

```python
os.listdir(r'..\data1\學長data\標記答案_最終答案')
```

## 讀取所有json檔案，存成csv

```python
##資料夾路徑
path='..\data1\學長data\標記答案_最終答案'
all_data_list=[]

#遍歷路徑中資料夾，存成list
for file in os.listdir(path):
    #判斷子檔案的後綴字詞是否為.json
    if file.endswith('.json'):
        #將path與file找到檔案字串進行串接
        load_data=os.path.join(path,file)
        #讀取json檔
        json_array=json.load(codecs.open(load_data, 'r', 'utf-8-sig'))
        #萃取key為推文的部分，存成json_tweets
        json_tweets= json_array['推文']
        context_list=[]
        answer_list=[]
        new_dict={}
        #迴圈依序挑出字典的value
        for i in range(1, len(json_tweets)+1):
            #分別萃取出字典1到最後的value，在挑出下層字典的留言內容的value與答案的value
            context_list.append(json_tweets[str(i)]['留言內容'])
            answer_list.append(json_tweets[str(i)]['答案'])
            #創建key為留言內容，value為由留言內容存成的list的字典
            new_dict['留言內容'] = context_list
            #創建key為答案，value為由答案存成的list的字典
            new_dict['答案']=answer_list
        #df為字典構成的dataframe   
        df = pd.DataFrame.from_dict(new_dict)
        #將dataframe新增到list(all_data_list)
        all_data_list.append(df)
#將list以concat的方式進行串接
all_data=pd.concat(all_data_list)      
all_data.to_csv (r"..\data1\學長data\標記答案_最終答案\context.csv", index = False, header=True,encoding='utf_8_sig')
```

## 讀取csv，存入all_data

```python
all_data = pd.read_csv(r"..\data1\學長data\標記答案_最終答案\context.csv")
all_data
```

## 將data轉成list(data_list)

```python
data_list= all_data['留言內容'].tolist()
data_list
```

```python
data_list_remove_space = []
for ele in data_list:
    j = str(ele).replace(' ', '')
    data_list_remove_space.append(j)

print(data_list_remove_space)
```

## CKIP斷詞、詞性分析、專有名詞辨識

```python
ws = WS("../data")
pos = POS("../data")
ner = NER("../data")
```

## 刪除空格(將空格以空字串代替)

```python
ws_results = ws(data_list_remove_space)
pos_results = pos(ws_results)
ner_results = ner(ws_results, pos_results)
```

```python
def print_word_pos_sentence(ws_results, pos_results):
    assert len(ws_results) == len(pos_results)
    for word, pos in zip(ws_results, pos_results):
        print(f"{word}({pos})", end="\u3000")
    print()
    return
    
for i, sentence in enumerate(data_list_remove_space):
    print()
    print(f"'{sentence}'")
    print_word_pos_sentence(ws_results[i],  pos_results[i])
    for entity in sorted(ner_results[i]):
        print(entity)
```

## 制造词典，拿出分词结果

```python
dictionary = corpora.Dictionary(ws_results )
dictionary.token2id
```

## 詞袋模型

#### 迴圈分词后的列表，把每一个分词结果拿出来
#### 然后将分词结果放到 dictionary.doc2bow 里面
#### dictionary.doc2bow(word) 里的 word 是每一句话的分词结果；dictionary.doc2bow(word)方法会把分词结果变成词袋模型展现形式
#### 然后通过列表生成式，将结果放到列表中

```python
words_bag = [dictionary.doc2bow(word) for word in ws_results ]
words_bag
```

```python
dic_wordbag={}
## for迴圈找出統計資料
for wb in words_bag:
    ## for迴圈，找出二維l
    for word,count in wb:
        if word in dic_wordbag.keys():
            dic_wordbag[word]+=count
        else:
            dic_wordbag[word]=count
        
```

```python
dic_wordbag
```

```python
dictList = [[]]
for key, value in dic_wordbag.items():
    #append list
    dictList[0].append((key,value))
dictList
```

```python
lst = [(1, 1), (2, 1), (3, 1), (4, 1), (5, 1), (6, 1), (7, 1), (8, 1)]

for x,y in lst:
    print(x,y)
    
```

```python
id_words = [[(dictionary[id], count) for id, count in line] for in dictList]
print(id_words)
```

```python
tfidf_model = models.TfidfModel(words_bag)
tfidf_model.save("tfidf_model.bin") 
fidf_model = models.TfidfModel.load("tfidf_model.bin") 
tfidf_vec = []
for i in words_bag:
    # tfidf 只是一个模型，需要通过模型，取出内容；i 是每一个 词袋模型的稀疏表示
    # 返回的 one_line_tfidf 也是稀疏表示
    one_line_tfidf = tfidf_model[i]
    tfidf_vec.append(one_line_tfidf)
print(tfidf_vec)
```

```python
tf_idf_words = [[(dictionary[id], count) for id, count in line] for line in tfidf_vec]
print(tf_idf_words)
```

```python
# Settings
seed = 666
sg = 0
window_size = 10
vector_size = 100
min_count = 1
workers = 8
epochs = 5
batch_words = 10000

model = word2vec.Word2Vec(
    ws_results,
    min_count=min_count,
    vector_size=vector_size,
    workers=workers,
    epochs=epochs,
    window=window_size,
    sg=sg,
    seed=seed,
    batch_words=batch_words
)

model.save('word2vec.model')

```
