```python
data = pd.read_csv(r"C:\Users\Robbie\Desktop\python\data1\自己爬蟲的小樣本\data1.csv")
data
```

```python
data_list= data['content'].tolist()
data_list
```

```python
ws = WS("./data")
pos = POS("./data")
ner = NER("./data")
```

```python
ws_results = ws(data_list)
print(ws_results)
```

```python
pos_results = pos(ws_results)
print(pos_results)
```

```python
ner_results = ner(ws_results, pos_results)
for name in ner_results:
    if(name != set()):
        print(name)
```
