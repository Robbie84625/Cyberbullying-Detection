```python
import pandas as pd
import jieba
import csv
import numpy
from gensim import corpora
from gensim import models
import jieba.posseg as pseg
```

```python
col_names = ['content']
# load dataset

data = pd.read_csv(r"C:\Users\Robbie\Desktop\python\data1\自己爬蟲的小樣本\data1.csv")
data.head()
```

```python
#讀取每一行內容
data=open(r"C:\Users\Robbie\Desktop\python\data1\自己爬蟲的小樣本\data1.csv",encoding="utf-8").read().split('\n')  

#建立儲存分詞的列表
Rs2=[] 

#讀取每一行分詞
for i in range(len(data)):
    result=[]
    seg_list = jieba.cut(data[i])
    #將該行分詞寫入總彙表形式的分詞列表
    for w in seg_list :
        result.append(w)
    Rs2.append(result)
#寫入CSV
file=open("data2.csv",'w',encoding="utf-8")
#定義寫入格式
writer = csv.writer(file)
#按行寫入
print(Rs2)
writer.writerows(Rs2)
#file.write(str(Rs))
file.close() 

```

## 詞袋模型

```python
# 制造词典，拿出分词结果
dictionary = corpora.Dictionary(Rs2)
#print(dictionary.token2id)
dictionary.token2id
```

```python
dic = {y: x for x, y in dictionary.token2id.items()}
dic
```

```python
# 迴圈分词后的列表，把每一个分词结果拿出来
# 然后将分词结果放到 dictionary.doc2bow 里面
# dictionary.doc2bow(word) 里的 word 是每一句话的分词结果；dictionary.doc2bow(word)方法会把分词结果变成词袋模型展现形式
# 然后通过列表生成式，将结果放到列表中
words_bag = [dictionary.doc2bow(word) for word in Rs2]
print(words_bag)
```

## TF-IDF

```python
tfidf_model = models.TfidfModel(words_bag)
tfidf_model.save("tfidf_model.bin") 
fidf_model = models.TfidfModel.load("tfidf_model.bin") 
tfidf_vec = []
for i in words_bag:
    # tfidf 只是一个模型，需要通过模型，取出内容；i 是每一个 词袋模型的稀疏表示
    # 返回的 one_line_tfidf 也是稀疏表示
    one_line_tfidf = tfidf_model[i]
    tfidf_vec.append(one_line_tfidf)
print(tfidf_vec)
```
