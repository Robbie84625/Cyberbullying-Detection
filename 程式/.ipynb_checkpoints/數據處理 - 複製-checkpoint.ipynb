## 讀取所有json檔案，存成csv

```python
##資料夾路徑
path='..\data1\學長data\標記答案_最終答案'
all_data_list=[]

#遍歷路徑中資料夾，存成list
for file in os.listdir(path):
    #判斷子檔案的後綴字詞是否為.json
    if file.endswith('.json'):
        #將path與file找到檔案字串進行串接
        load_data=os.path.join(path,file)
        #讀取json檔
        json_array=json.load(codecs.open(load_data, 'r', 'utf-8-sig'))
        #萃取key為推文的部分，存成json_tweets
        json_tweets= json_array['推文']
        context_list=[]
        answer_list=[]
        new_dict={}
        #迴圈依序挑出字典的value
        for i in range(1, len(json_tweets)+1):
            #分別萃取出字典1到最後的value，在挑出下層字典的留言內容的value與答案的value
            context_list.append(json_tweets[str(i)]['留言內容'])
            answer_list.append(json_tweets[str(i)]['答案'])
            #創建key為留言內容，value為由留言內容存成的list的字典
            new_dict['留言內容'] = context_list
            #創建key為答案，value為由答案存成的list的字典
            new_dict['答案']=answer_list
        #df為字典構成的dataframe   
        df = pd.DataFrame.from_dict(new_dict)
        #將dataframe新增到list(all_data_list)
        all_data_list.append(df)
#將list以concat的方式進行串接
all_data=pd.concat(all_data_list)      
all_data.to_csv (r"..\data1\學長data\標記答案_最終答案\context.csv", index = False, header=True,encoding='utf_8_sig')
```

## 刪除中文標點符號

```python
data_remove_punctuation=[]
for word in RemoveUrl:
    for i in punctuation:
        word=str(word).replace(i,'')
    data_remove_punctuation.append(word)
data_remove_punctuation = [x.strip() for x in data_remove_punctuation if x.strip()!='']
```

## 刪除英文標點符號

```python
data_remove_punctuation = [''.join(letter for letter in word if letter not in string.punctuation) for word in data_remove_punctuation]
data_remove_punctuation = [x.strip() for x in data_remove_punctuation if x.strip()!='']
```

## 刪除空格(將空格以空字串代替)

```python
data_list_remove_space = []
for ele in data_remove_punctuation:
    j = str(ele).replace(' ', '')
    data_list_remove_space.append(j)
data_list_remove_space
```

## 詞袋模型

#### 迴圈分词后的列表，把每一个分词结果拿出来
#### 然后将分词结果放到 dictionary.doc2bow 里面
#### dictionary.doc2bow(word) 里的 word 是每一句话的分词结果；dictionary.doc2bow(word)方法会把分词结果变成词袋模型展现形式
#### 然后通过列表生成式，将结果放到列表中

```python
words_bag = [dictionary.doc2bow(word) for word in remove_stw]
```

```python
id_words = [[(dictionary[id], count) for id, count in line] for line in words_bag]
```
